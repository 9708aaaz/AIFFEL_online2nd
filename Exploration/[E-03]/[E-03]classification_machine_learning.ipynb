### (4) train, test 데이터 분리

```python
# 모델 학습과 테스트용 문제지와 정답지를 준비해 봅시다.
# X_train, X_test, y_train, y_test를 생성하는 방법을 참고해 보세요.

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(digit_data, 
                                                    digit_label, 
                                                    test_size=0.2, 
                                                    random_state=32)

print('X_train 개수: ', len(X_train),', X_test 개수: ', len(X_test))
```

### (5) 다양한 모델로 학습시켜보기

```python
# 학습데이터 X_train, y_train 을 활용해 분류기 모델을 만들어 봅시다. 어떤 모델이 가장 좋은 성능을 보일까요?
```

```python
# Decision Tree 사용해 보기
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

decision_tree = DecisionTreeClassifier(random_state=32) 
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test) 

print(classification_report(y_test, y_pred))
```

```python
# Random Forest 사용해 보기
from sklearn.ensemble import RandomForestClassifier 

random_forest = RandomForestClassifier(random_state=32) 
random_forest.fit(X_train, y_train) 
y_pred = random_forest.predict(X_test) 

print(classification_report(y_test, y_pred)) 
```

```python
# SVM 사용해 보기
from sklearn import svm

svm_model = svm.SVC()
svm_model.fit(X_train, y_train) 
y_pred = svm_model.predict(X_test)

print(classification_report(y_test, y_pred)) 
```

```python
# SGD Classifier 사용해 보기
from sklearn.linear_model import SGDClassifier

sgd_model = SGDClassifier() 
sgd_model.fit(X_train, y_train) 
y_pred = sgd_model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
# Logistic Regression 사용해 보기
from sklearn.linear_model import LogisticRegression 

logistic_model = LogisticRegression(max_iter=5000)
logistic_model.fit(X_train, y_train) 
y_pred = logistic_model.predict(X_test) 

print(classification_report(y_test, y_pred))
```

#### 회고 : 로지스틱 회귀에서만 'ConvergenceWarning: lbfgs failed to converge (status=1)' 이런 에러가 떠서, 제시된 해결책대로 max_iter 값을 늘려주었다.

#### support : 각 라벨의 실제 샘플 개수 
#### macro avg : 단순 평균, 샘플 개수의 불균형을 고려하지 않고 각 클래스 별로 동일한 가중치 부여, 그러나 소수 클래스에 대한 성능이 낮을 때(데이터가 불균형할 때) 더 큰 패널티 발생
#### weighted avg : 가중평균, 각 클래스에 속하는 표본의 개수로 가중평균 내서 계산하는 방법, 샘플 개수의 불균형을 고려함

#### 게다가 테스트 데이터 예측 결과가 불균형하지 않기 때문에 precision에 대한 macro avg로 판단할 수 있다.

# 총 회고 

다양한 분류를 위한 머신러닝 모델을 활용하면서, 지난 토요일의 회귀 모델 프로젝트와 유사한 부분도 있고 다른 부분도 있어서 이해도를 높일 수 있었다.   
그렇지만 아직 다양한 분류 모델에서 왜 지표가 나르게 나타나는지, 각 지표들의 차이가 무엇인지는 잘 이해되지 않는다. 특히 각 데이터셋에 대해 어떤 지표가 적절할지 판단하기 위해 평가 지표에 대한 다양한 사이트를 검색해봤는데, 어렴풋이 이해되는 듯 해도 막상 판단하기엔 어려웠다.

로지스틱 회귀에서 에러가 났는데, 에러와 함께 max_iter를 늘려보라는 해결법도 같이 출력되었다. 처음엔 지난 프로젝트에서 회귀계수가 크면 너무 느렸던게 생각나 10부터 시작했으나 1000까지 늘려도 에러가 계속 났다. 결국 5000으로 늘리자 에러가 해결되었다. 

또한 SVM 모델을 돌릴 땐, 처음 코드를 짤 때는 안 나왔던 UndefinedMetricWarning 에러가 났는데, 검색해보니 f1-score만 0으로 출력하는 것 외에는 전체 모델에 별로 영향을 미치지 않았다. 그래서 에러 메세지를 무시하는 코드를 넣어주었다.  

파이썬 > 넘파이 > 판다스 > 머신러닝으로 넘어가면서 새로운 용어가 많이 등장했는데, 머신러닝 용어는 검색해봐도 아직 잘 이해되지 않는다. 게다가 머신러닝에 그 전에 배웠던 부분이 뒤섞이니 머리가 더 복잡해지는 것 같다. 머신 러닝 용어에 대한 책을 찾아 기초를 다져야될 것 같다.
